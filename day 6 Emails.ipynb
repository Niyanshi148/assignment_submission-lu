```python
server = smtplib.SMTP("smtp.gmail.com",587)
server.ehlo()
server.starttls()
server.login("sender's mail id","sender's mail id password")
server.sendmail("sender's mail id",
                "reciever's mail id",
               "hey ,this mail is send by python program")
server.close()
```
